```python
% matplotlib notebook
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
#pd.get_option("display.max_rows", 10)
```

```python
colnames = ['Ref', 'AvgSupplyDelay']
new_purchase_data = pd.read_csv('Lokad_PurchaseOrders.csv').merge(pd.read_csv('Lokad_AverageSupplyDelay.csv', 
                                                            names = colnames, header = None), on = 'Ref', right_index = True)

new_purchase_data
```

```python
# For the Avarage daily usage of product is calculated using the Lokad_Orders dataset 

lokad_orders = pd.read_csv('Lokad_Orders.csv')
lokad_orders
```

```python
# Average daily usage sorted on the basis of Reference of the product.
#def avg_usage():
avg_daily_usage = lokad_orders.groupby('Ref')['Quantity'].mean().astype(int)
    #return avg_daily_usage
avg_daily_usage.to_csv('Quantity.csv')
```

```python
colnames = ['Ref', 'DailyUsage']
data = pd.read_csv('Quantity.csv', names = colnames, header = None)

data
```

```python
new_data = new_purchase_data.merge(data, on = 'Ref', right_index = True)
new_data
```

```python
new_data['Demand'] = new_data.AvgSupplyDelay * new_data.DailyUsage
Avg_lead_demand = new_data.groupby('Ref')['Demand'].mean()
```
