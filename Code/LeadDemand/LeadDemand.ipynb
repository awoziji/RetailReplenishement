## Reading the csv files

```python
% matplotlib notebook
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
#pd.get_option("display.max_rows", 10)
```

```python
colnames = ['Ref', 'AvgSupplyDelay']
new_purchase_data = pd.read_csv('temp1.csv', index_col = 0)

new_purchase_data
```

## Loading the data from Orders.csv

```python
# For the Avarage daily usage of product is calculated using the Lokad_Orders dataset 

lokad_orders = pd.read_csv('Lokad_Orders.csv', index_col = 0)
lokad_orders
```

```python
# Average daily usage sorted on the basis of Reference of the product.
#def avg_usage():
avg_daily_usage = lokad_orders.groupby('Ref')['Quantity'].mean().astype(int)
    #return avg_daily_usage
avg_daily_usage.to_csv('Quantity.csv')
```

```python
# Reading thr reference id and Daily usage of the product from the Quantity.csv

colnames = ['Ref', 'DailyUsage']
data = pd.read_csv('Quantity.csv', names = colnames, header = None)

data
```

```python
new_data = new_purchase_data.merge(data, on = 'Ref', right_index = True)
new_data.to_csv('temp1.csv')
new_data
```

### Lead Demand calculation 
#### The projected level of demand for a product from consumers during its lead time from the supplier to the retailer
#### Lead Demand = Lead Time(LT in days) * Avg Daily Usage(of products in units)

```python
new_data['Demand'] = new_data.LeadTime * new_data.DailyUsage
new_data['Demand']
```

```python
Avg_demand = new_data.groupby('Ref')['Demand'].mean()
Avg_demand
```

```python
Avg_demand.to_csv('temp_Davg.csv')

colnames = ['Ref','Davg']

new_df = pd.read_csv('temp_Davg.csv', names = colnames, header = None)

new_data.merge(new_df, on = 'Ref', right_index = True).to_csv('temp1.csv')

df_new = pd.read_csv('temp1.csv', index_col = 0)
df_new
```

### Variance of Demand

```python
var_LD = new_data.groupby('Ref')['Demand'].var()
```

```python
var_LD.fillna(0).astype(int)
```

## Standard deviation of demand

```python
# The standard deviation is calculated for it's usage in the safety stock level calculation.
```

```python
sigma_LD = new_data.groupby('Ref')['Demand'].std().fillna(0).astype(int)
sigma_LD.to_csv('temp.csv')
colname = ['Ref', 'sigmaD']
df = pd.read_csv('temp.csv', names = colname, header = None)
df
```

```python
df_new.merge(df, on = 'Ref', right_index = True).to_csv('temp1.csv')
df_new
```

### Lead Demand Forecast

```python
# Decision Tree Algorithm

from sklearn.ensemble import ExtraTreesClassifier
from sklearn.metrics import accuracy_score

# define the dataset
X = df_new[['LeadTime', 'DailyUsage']]
y = df_new['Demand']

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 0)
# fit an Extra Trees model to the data
model = ExtraTreesClassifier()
model.fit(X, y)
# display the relative importance of each attribute
print(model.feature_importances_)
```

```python
model.score(X_test, y_test) * 100
```
