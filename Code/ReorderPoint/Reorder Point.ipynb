## Reorder Point 

#### Inventory level of an item which signals the need for a replenishment order. 
#### The reorder point is classically viewed as the sum of the lead demand plus the safety stock
##### Reorder point (RP) = Safety Stock Level (SSL) + Lead Demand(LD)

```python
%matplotlib notebook
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
data = pd.read_csv('temp1.csv', index_col = 0)
data
```

```python
colname = ['SSL']
df = pd.read_csv('temp_SSL.csv', names = colname, header = None, index_col = 0)
df
```

```python
RP = (data.Demand + df.SSL).fillna(0).astype(int)
```

```python
## Stroing of the Reorder Point(RP) in the form of csv for all the products taken into account.
```

```python
RP.to_csv('temp_RP.csv')
```

```python
colname = ['Reorder Point']
df_RP = pd.read_csv('temp_RP.csv', names = colname, header = None, index_col = 0)
df_RP
```

```python
data.merge(df_RP, left_index = True, right_index = True).to_csv('temp1.csv')
```
