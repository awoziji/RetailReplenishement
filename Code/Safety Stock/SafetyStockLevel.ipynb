```python
% matplotlib notebook
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('temp1.csv', index_col = 0)
data
```

### Calculation of sigmaD- sqaure * Lead Time

```python
square = np.power(data.sigmaD, 2).fillna(0).astype(int)
```

```python
temp1 = data.LeadTime * square
temp1
```

### Calculation of sigmaLT * Davg

```python
temp2  = (data.sigmaLT * data.Davg).astype(int)
temp2
```

```python
temp2_square = abs(np.power(temp2, 2).fillna(0).astype(int))
temp2_square
```

### Safety Stock Level being calculated for all the products.

```python
# The Safety STock Level is being represented as SSL in the project. It marks the minimum buffer stock that must be kept in hold
```

```python
data['temp'] = temp2 + temp2_square
data['temp']
##SSL = np.sqrt(temp)
```

```python
SSL = np.sqrt(data['temp']).fillna(0).astype(int)
```

#### Storing of the SSL calculated in the output file.

```python
SSL.to_csv('temp_SSL.csv')
```
